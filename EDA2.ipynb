```python
categorical_cols_label = ['education', ' workclass', 'marital_status', 'occupation', 'relationship', 'native_country', 'race']
```

```python
categorical_cols_one_hot = ['sex', 'income']
```

```python
# Import label encoder 
from sklearn import preprocessing 

# label_encoder object knows 
# how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 

# Encode labels in column. 
data['education']= label_encoder.fit_transform(data['education']) 
data['workclass']= label_encoder.fit_transform(data['workclass']) 
data['occupation']= label_encoder.fit_transform(data['occupation']) 
data['relationship']= label_encoder.fit_transform(data['relationship']) 
data['native_country']= label_encoder.fit_transform(data['native_country']) 
data['race']= label_encoder.fit_transform(data['race']) 
data['marital_status']= label_encoder.fit_transform(data['marital_status']) 
```

```python
data
```

```python
#Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Apply one-hot encoding to the categorical columns
one_hot_encoded = encoder.fit_transform(data[categorical_cols_one_hot])

one_hot_data = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_cols_one_hot))

data_encoded = pd.concat([data, one_hot_data], axis=1)

data_encoded = data_encoded.drop(categorical_cols_one_hot, axis=1)

print(f"Encoded Employee data : \n{data_encoded}")
```
